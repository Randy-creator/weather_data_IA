```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
# Configurations
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
# Load the dataset
data_path = Path("dags/data/historical_weather/historical_summary.csv")  # Adjust if necessary
df = pd.read_csv(data_path)
```

```python
# Preview the data
print("Preview of the dataset:")
df.head()
```

```python
# General info and missing values
print("General information:")
df.info()
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Descriptive statistics
print("\nDescriptive statistics:")
display(df.describe())
```

```python
# average temperature per city
plt.figure()
sns.boxplot(data=df, x="city", y="average temperature")
plt.title("average Temperature per City")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# average humidity per city
plt.figure()
sns.barplot(data=df, x="city", y="humidity", estimator='mean', errorbar=None)
plt.title("average umidity per City")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create a date column for temporal analysis
df["month_num"] = pd.to_datetime(df["months"], format="%B").dt.month
df["month_str"] = df["month_num"].apply(lambda x: f"{x:02d}")
df["date"] = pd.to_datetime(df["year"].astype(str) + "-" + df["month_str"])
```

```python
# evolution of average temperature over time per city
plt.figure()
for city in df["city"].unique():
    sub = df[df["city"] == city].sort_values("date")
    plt.plot(sub["date"], sub["average temperature"], label=city)

plt.title("Evolution of Average Temperature Over Time")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# wind speed per city
plt.figure()
sns.boxplot(data=df, x="city", y="wind speed")
plt.title("Wind Speed per City")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Monthly average metrics across all cities
monthly_avg = df.groupby("months").mean(numeric_only=True).reindex([
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
])
```

```python
monthly_avg[["average temperature", "humidity", "wind speed"]].plot(marker='o')
plt.title("Monthly Averages Across All Cities")
plt.ylabel("Average Values")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
